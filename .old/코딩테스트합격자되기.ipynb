607 주식가격 예측하기

```python
prices = [1,1,1,1,1,99]

def sol607(prices):
	n = len(prices)
	answer = [0]*n

	stack = [0]
	for i in range(1,n):
		while stack and prices[i] < prices[stack[-1]]:
			j = stack.pop()
			answer[j] = i-j
		stack.append(i)
	while stack:
		j = stack.pop()
		answer[j] = n-1-j

	return answer
```

- 스택에 값의 인덱스를 쌓는데, 처음에 생각나는 방법은 인덱스의 값을 저장하고 있는 배열을 생성하거나 딕셔너리 형태로 만드는것
- 근데 그렇게 안하고 0으로 초기화된 배열을 만들고 거리를 계산하는 방법이 있다. -> 여기에 익숙해져야..
